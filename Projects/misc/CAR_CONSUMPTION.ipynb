# Models: Regression Problem - predict horsepower for missing values



```python
from sklearn.linear_model import ElasticNet;
from sklearn.linear_model import LinearRegression;
from sklearn.svm import LinearSVR;
from sklearn.ensemble import RandomForestRegressor;
from sklearn.svm import SVR;
import pickle;
from sklearn.model_selection import LeaveOneOut;
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score;
```

```python
lin_reg = LinearRegression();
lin_reg_elast = ElasticNet(max_iter=90000, l1_ratio=1, positive=False, tol=0.000001, selection='random');#para
LSVR = LinearSVR(max_iter=90000);#non-para
SVr = SVR(kernel="linear", C=15);#non-para
RFR = RandomForestRegressor(n_estimators=60);#non-para
```

```python
x = analysis_2_withoutNAN.drop(["mpg", "displacement", "weight", "horsepower"], axis="columns");
y = analysis_2_withoutNAN.loc[:, "horsepower"];
```

# Model Validation Techniques:
### Cross Validation:
####     -KFold: (Low Bias at KFold = 10)
####     -LOOCV (Leave-One-Out-Cross-Validation): Most Accurate, omputationally expensive

```python
cv = LeaveOneOut();#Because my data set < 500 rows
```

```python
lin_reg_ = np.absolute(np.mean(cross_val_score(lin_reg, x, y, scoring="neg_mean_absolute_error", cv=cv, n_jobs=-1)));
lin_reg_elast_ = np.absolute(np.mean(cross_val_score(lin_reg_elast, x, y, scoring="neg_mean_absolute_error", cv=cv, n_jobs=-1)));
LSVR_ = np.absolute(np.mean(cross_val_score(LSVR, x, y, scoring="neg_mean_absolute_error", cv=cv, n_jobs=-1)));
SVr_ = np.absolute(np.mean(cross_val_score(SVr, x, y, scoring="neg_mean_absolute_error", cv=cv, n_jobs=-1)));
RFR_ = np.absolute(np.mean(cross_val_score(RFR, x, y, scoring="neg_mean_absolute_error", cv=cv, n_jobs=-1)));
```

```python
print(lin_reg_, lin_reg_elast_, LSVR_, SVr_, RFR_);#Sure that the smallest here is good
```

```python
results = np.array([lin_reg_, lin_reg_elast_, LSVR_, SVr_, RFR_]);
```

```python
results[np.where(results == results.min())[0][0]]
```

# Phone Specs DataSet

```python
import numpy as np;
import pandas as pd;
import math;
```

```python
phone_df_1 = pd.read_csv("../Assignments/datasets/phone_price/train.csv");
phone_df_1.drop(['price_range'], axis=1, inplace=True);
```

```python
phone_df_2 = pd.read_csv("../Assignments/datasets/phone_price/test.csv");
phone_df_2.drop(["id"], inplace=True, axis=1);
```

```python
merged = pd.concat([phone_df_1, phone_df_2.iloc[0:, :]], axis="rows");
phone_df = merged;
merged
```

```python
#Drop all rows with pnly 50% data
phone_df.dropna(thresh=math.floor(len(phone_df.columns)/2), inplace=True);
```

```python
#Shuffle the dataSet
phone_df = phone_df.sample(frac=1).reset_index(drop=True);
```

```python
phone_df.columns
```

```python
analysis = phone_df.sample(frac=1).iloc[:, :];
analysis
```

```python
analysis.reset_index(drop=True, inplace=True);
```

```python
analysis = analysis.loc[:, ["ram", "clock_speed", "battery_power", "dual_sim", "four_g", "n_cores", "sc_w", "sc_h", "px_width", "px_height", "touch_screen", "wifi"]].sort_values(by=["clock_speed", "ram"])
```

```python
analysis.reset_index(drop=True, inplace=True);
```

```python
analysis
```

```python
#Give a numerical value for the phone's performace
def performance(row):
    result = ((row.ram*row.clock_speed*row.n_cores)+row.battery_power);
    return np.log(result);
```

```python
analysis["performance"] = analysis.apply(lambda row: performance(row), axis=1);
```

```python
maximum_performance = analysis["performance"].max()+1.5;
```

```python
analysis["performance"] = analysis["performance"].apply(lambda value: value/maximum_performance);
```

```python
for_model = analysis.drop(["sc_w", "sc_h", "px_width", "px_height", "touch_screen"], axis="columns");
```

```python
from sklearn.model_selection import KFold;
from sklearn.linear_model import LinearRegression;
from sklearn.tree import DecisionTreeRegressor;
from sklearn.ensemble import RandomForestRegressor;
from sklearn.svm import LinearSVR;
```

```python
lin_reg = LinearRegression();
DTR = DecisionTreeRegressor();
RFR = RandomForestRegressor(n_estimators=10);
LSV = LinearSVR(max_iter=500);
```

```python
def get_accuracy(model, x_train, x_test, y_train, y_test):
    model.fit(x_train, y_train);
    return model.score(x_test, y_test);
```

```python
kf = KFold(n_splits=80);
x = for_model.drop(["performance"], axis="columns");
y = for_model["performance"];
```

```python
#Increase the weight of wifi for a phone by increase the scale
def times_50(number):
    print(number);
    return lambda x: x*50;
```

```python
for_model["wifi"].astype("uint");
for_model["wifi"] = for_model["wifi"].apply(lambda x: x/50);
for_model
```

```python
lin_reg_ = [];
DTR_ = [];
RFR_ = [];
LSV_ = [];
for train_index, test_index in kf.split(x, y):
    try:
        x_train, x_test, y_train, y_test = x.iloc[train_index, :], x.iloc[test_index, :], y.iloc[train_index], y.iloc[test_index];
        lin_reg_.append(get_accuracy(lin_reg, x_train, x_test, y_train, y_test));
        DTR_.append(get_accuracy(DTR, x_train, x_test, y_train, y_test));
        RFR_.append(get_accuracy(RFR, x_train, x_test, y_train, y_test));
        LSV_.append(get_accuracy(LSV, x_train, x_test, y_train, y_test));
    except:
        continue
```

```python
print(np.array(lin_reg_).mean(), np.array(DTR_).mean(), np.array(RFR_).mean(), np.array(LSV_).mean())
```
